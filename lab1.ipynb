```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0.0, 50, 0.1)
y = x
fig = plt.figure()
graph = plt.plot(x, y)
plt.title('Draw a line')
plt.xlabel('x - axis')
plt.ylabel('y - axis')

plt.show()
```

```python
x1 = [10, 20, 30]
y1 = [20, 40, 10]
y2 = [40, 10, 30]
fig = plt.figure()
ax = fig.add_subplot(111)
line1 = ax.plot(x1, y1, color='blue', linewidth=3, label='line1-width-3')
line2 = ax.plot(x1, y2, color='red', linewidth=5, label='line2-width-5')
plt.title('TITLE')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
ax.legend()

plt.show()

```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
line1 = ax.plot(x1, y1, color='blue', linewidth=3, linestyle='dotted', label='line1-dotted')
line2 = ax.plot(x1, y2, color='red', linewidth=5, linestyle='dashed', label='line2-dashed')
ax.legend()
plt.title('Title')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.show()
```
