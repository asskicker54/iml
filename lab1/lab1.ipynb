```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0.0, 50, 0.1)
y = x
fig = plt.figure()
graph = plt.plot(x, y)
plt.title('Draw a line')
plt.xlabel('x - axis')
plt.ylabel('y - axis')

plt.show()
```

```python
x1 = [10, 20, 30]
y1 = [20, 40, 10]
y2 = [40, 10, 30]
fig = plt.figure()
ax = fig.add_subplot(111)
line1 = ax.plot(x1, y1, color='blue', linewidth=3, label='line1-width-3')
line2 = ax.plot(x1, y2, color='red', linewidth=5, label='line2-width-5')
plt.title('TITLE')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
ax.legend()

plt.show()

```

```python
from matplotlib.ticker import FormatStrFormatter
x = np.array(['2016-10-03', '2016-10-04', '2016-10-05', '2016-10-06', '2016-10-07'])
y = np.array([772.5, 776.4, 776.5, 776.8, 775])
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot()
ax.minorticks_on()
ax.grid(which='major', lw=2, color='r')
ax.grid(which='minor', linestyle='--')
ax.set_ylim([772.5, 777.0])
ax.yaxis.set_major_formatter(FormatStrFormatter("%0.1f"))
ax.plot(x, y, color='r', marker='o', markerfacecolor='r')
plt.show()
```

```python

import matplotlib.ticker as ticker
x1 = np.arange(0.25, 4, 0.25)
y1 = np.sin(x1/2) * 4
y2 = np.log(x1**0.6)
x = np.random.rand(3, 3)
y = np.random.rand(3, 3)
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111)
plt.title(' ')
ax.set_xlabel('OX')
ax.set_ylabel('OY')
ax.grid(linestyle='--')
ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(0.25))
ax.xaxis.set_minor_formatter(ticker.FormatStrFormatter('%.2f'))
ax.yaxis.set_major_locator(ticker.MultipleLocator(1))
ax.yaxis.set_minor_locator(ticker.MultipleLocator(0.25))
ax.yaxis.set_minor_formatter(ticker.FormatStrFormatter('%.2f'))
g1 = ax.plot(x1, y1, color='r', label=' ')
g2 = ax.plot(x1, y2, color='b', label=' ')
ax.scatter(x, y, marker='o', c='k')
ax.legend()
plt.show()
```

```python
import math
x = np.arange(-10, 10)
number = (x**2 + 1) * np.exp(np.abs(x)/-10)
base = 1 + np.tan(1/(1 + np.sin(x)**2))
y = np.log(number) / np.log(base)
plt.plot(x, y)
plt.show()
```

```python
x = ['Java', 'Python', 'PHP', 'JS', 'C#', 'C++']
y = [22.5, 17.5, 9, 8, 7, 6]
fig = plt.figure()
ax = fig.add_subplot(111)
ax.minorticks_on()
ax.grid(which='major', c='r')
ax.grid(which='minor', linestyle=':')
barlist = ax.bar(x, y)
barlist[0].set_color('r')
barlist[1].set_color('k')
barlist[2].set_color('g')
barlist[3].set_color('b')
barlist[4].set_color('y')
barlist[5].set_color('c')
plt.title('Pop of programming languages')
plt.show()
```

```python
x = ['Java', 'Python', 'PHP', 'JS', 'C#', 'C++']
y = [22.5, 17.5, 9, 8, 7, 6]
fig = plt.figure()
ax = fig.add_subplot(111)
ax.minorticks_on()
ax.grid(which='major', c='r')
ax.grid(which='minor', linestyle=':')
bars = ax.bar(x, y, color='b')
for bar in bars:
    value = bar.get_height()
    text = f'{value * 100000}'
    text_x = bar.get_x() + bar.get_width() / 2
    text_y = bar.get_y() + value
    ax.text(text_x, text_y, text, ha='center',color='k',size=12)
plt.title('Pop of programming languages')
plt.show()
```
